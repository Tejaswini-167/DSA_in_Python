* It returns the expression back to the function
* The code after the return statement is not executed
* If there is no return value the function returns none

```python
def add(a,b):
    sum=a+b
    print(sum)

add(2,4)
```

```python

c=add(2,4)
```

### Write a puthon function to print even numbers in a given list

* #### input : [1,2,3,4,5,6]
  
* #### output : 2,4,6

```python

def even_num(list):
    for i in list:
        if i % 2==0:
            print(i, end=" ")

list=[1,2,3,4,5,6]
```

```python
even_num(list)
```

### Write a python function that takes a list and returns a new list with unique elements of first list
* #### input : [1,2,3,1,2,4]
* #### output : 1,2,3,4

```python
def unique_list(list):
    new_list=[]
    for i in list:
        if i not in new_list:
            new_list.append(i)
    return new_list


list=[1,2,3,1,2,4]
```

```python
unique_list(list)
```
