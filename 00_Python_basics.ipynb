```python
# FIND MAXIMUM IN A LIST
marks=[90,30,100,50,80,95]
max_i=marks[0]
for i in marks:
    if i>max_i:
        max_i=i

print( max_i)
```

```python
# GRADING SYSTEM
marks = int(input("Enter your marks: "))

if marks >= 60 and marks < 70:
    print("Grade D")
elif marks >= 70 and marks < 80:
    print("Grade C")
elif marks >= 80 and marks < 90:
    print("Grade B")
elif marks >= 90:
    print("Grade A")
else:
    print("Fail")
```
