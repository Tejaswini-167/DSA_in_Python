## Conditional Statement

#### Conditional statements in Python are used to execute certain blocks of code based on specific conditions. 
* if statement
* if else statement
* if elif else statement
* nested if statement

## if statements
#### If statement is the simplest form of a conditional statement. It executes a block of code if the given condition is true.

```python
age = 20
if age >= 18:
    print("Eligible to vote.")
```

## If else Conditional Statement
#### If Else allows us to specify a block of code that will execute if the condition(s) associated with an if or elif statement evaluates to False.

```python
age = 10
if age <= 12:
    print("Travel for free.")
else:
    print("Pay for ticket.")
```

## if elif else Statement
#### if elif else statement in Python allows us to check multiple conditions, providing a way to execute different blocks of code based on which condition is true. 

```python
age = 25

if age <= 12:
    print("Child.")
elif age <= 19:
    print("Teenager.")
elif age <= 35:
    print("Young adult.")
else:
    print("Adult.")
```

```python
a=10
b=5
c=2
if a>b and a>c:
    print("greater",+ a)
elif b>c and b>a:
    print("greater",+ b) 
else:
    print("greater",+ c)
    
```

## Nested if..else Conditional Statement
#### Nested if..else means an if-else statement inside another if statement. We can use nested if statements to check conditions within conditions.

```python
age = 70
is_member = True

if age >= 60:
    if is_member:
        print("30% senior discount!")
    else:
        print("20% senior discount.")
else:
    print("Not eligible for a senior discount.")
```

# break continue and passs
* these statements are used to alter the flow of a program
* ####  break: breaks the flow of program once this condition is hit
* #### continue: its skips the paticular iteration
* #### pass:to avoid syntax error

### break statement: Stops the loop immediately.

#### Use cases 

* Exit loop on success
* Stop when condition met

```python
for i in range(1, 10):
    if i == 5:
        break
    print(i,end= ' ')
```

### continue statement: Skips current iteration.




#### break vs continue:

* break exits loop
* continue skips one step

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i, end= ' ')
```

```python
# Number guessing game

secret = 7

while True:
    guess = int(input("Guess the number: "))

    if guess == secret:
        print("Correct guess")
        break
    else:
        print("Try again")
```

#### 2. Keep asking age until valid age above 0


```python
while True:
    age=int(input(" Enter your age: "))

    if age>0:
        print("your age is: ", age)
        break
    else:
        print("enter a valid age")
        
            
```

#### 3. Sum numbers until user enters 0


```python
total = 0
num = int(input("Enter a number: "))

while num != 0:
    total = total + num
    num = int(input("Enter a number: "))

print("Sum is:", total)
```

#### 4. Skip multiples of 5 from 1 to 50


```python
for i in range(1,51):
    
    if i%5 == 0:
        continue
    else:
        print(i,end=" ")
        i=i+1
```

#### 5. Stop loop when user enters “exit”

```python
while True:
    text = input("Type something: ")

    if text == "exit":
        print("Loop stopped")
        break

    print("You typed:", text)
```
