# challange: find if a string is a palindrome
* a strign is said to be palindrome if the reverse of the string is the same as string
* for example, "radar" is a palindrome 

```python
string=input()
if string[:] == string[::-1]:
    print(f"the given string {string} is a palindrome")
else:
    print(f"the given string {string} is not a palindrome")
```

```python
import string

def set_operation(sent1, sent2):
    ''' 
    input: sent1, sent2 - two sentences taken as inputs
    output: return the sum of the total number of unique words from both sentences.
    '''
    
    # Helper function to clean the sentence and split into words
    def get_words(sentence):
        # Convert sentence to lowercase and remove punctuation
        sentence = sentence.lower().translate(str.maketrans('', '', string.punctuation))
        words = sentence.split()
        return words
    
    # Get words from both sentences
    words1 = get_words(sent1)
    words2 = get_words(sent2)

    # Create a set from words in each sentence to count unique words
    unique_count1 = len(set(words1))  # set removes duplicates from words1
    unique_count2 = len(set(words2))  # set removes duplicates from words2

    # Return the sum of the unique word counts from both sentences
    return unique_count1 + unique_count2
```
